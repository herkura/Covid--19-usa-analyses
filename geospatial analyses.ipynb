```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium

print('Folium installed and imported!')
```

```python
lat_usa=39.381266
lng_usa= -97.922211
```

```python
# creating a map and displaying it
usa_map = folium.Map(location=[lat_usa , lng_usa],zoom_start=4)
usa_map
```

```python
#superimposing the data with the map
cases = folium.map.FeatureGroup()
for lat,lng in zip(df.lat,df.long):
    cases.add_child( folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
      
usa_map.add_child(cases)
```

```python
from folium import plugins
usa_map = folium.Map(location=[lat_usa , lng_usa],zoom_start=4)
cases =  plugins.MarkerCluster().add_to(usa_map)
for lat,lng in zip(df.lat,df.long):
    folium.Marker(
    location=[lat,lng],
    icon=None,
    
    ).add_to(cases)
usa_map    
```
